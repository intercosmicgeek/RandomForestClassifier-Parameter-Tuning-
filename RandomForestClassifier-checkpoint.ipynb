## RandomForestClassifier Algorithm using seaborn's Bulit-in Titanic Dataset
### This problem solving involves
#### DataPreprocessing
#### Hyper-Parameter Optimization to avoid OverFitting

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

# Data-Preprocessing

## In above dataset, few features has same information like 'survived' and 'alive' ,
## 'sex' and 'who' we can use "who" because it has additional data compared to "sex"
## and here in this case,the town/city does not even make sense 
## and also the "adult_male" column also don't make much difference
### we can drop those columns/Features

```python
df=df.drop(['alive','sex','adult_male','embark_town'],axis=1)
```

```python
df.head(2)
```

```python
df.shape
```

## Handling Null Values

```python
df.isnull().sum()
```

```python
# Filling the null values in age with the mean of all the rows in age
df['age']=df['age'].fillna(df['age'].mean())
```

## Dataset has 891 rows and the 'deck' column contains more than 75% of null values
## It is good to remove 'deck' feature from our Dataset

```python
df=df.drop(['deck'],axis=1)
```

```python
df.head(1)
```

```python
print("unique values of embarked: ",df['embarked'].unique())
print("mode is: ",df['embarked'].mode())
```

```python
## filling null values of 'embarked' feature with its mode
df['embarked']=df['embarked'].fillna('S')
```

```python
df.isnull().sum()
```

## Encoding with sklearn's Labelencoder

```python
df.info()
```

```python
#importing libraries
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()
```

```python
df['class']=encoder.fit_transform(df['class'])
```

```python
df['embarked']=encoder.fit_transform(df['embarked'])
```

```python
df['who']=encoder.fit_transform(df['who'])
df['alone']=encoder.fit_transform(df['alone'])
```

```python
df.info()
```

## All the features were encoded and there is no object datatypes

```python
#importing libraries for visualization
import matplotlib.pyplot as plt

df.describe().plot()
```

```python
sns.pairplot(df,hue='survived')
```

```python
from sklearn.ensemble import RandomForestClassifier

classifer=RandomForestClassifier()

#separating independent and dependent features
x=df.drop(['survived'],axis=1)
y=df['survived']
```

```python
x.head(1)
```

```python
y.head(1)
```

```python
#spliting Training and Testing data 
from sklearn.model_selection import train_test_split

X_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
#without parameter tuning
classifer.fit(X_train,y_train)
print("Train accuracy: ",classifer.score(X_train,y_train))
```

```python
y_pred=classifer.predict(x_test)
```

```python
#Accuracy score and confusion matrix
from sklearn.metrics import accuracy_score,confusion_matrix

print("Accuracy Score of Test Data: ",accuracy_score(y_test,y_pred))
print("\nConfusion Matrix")
print(confusion_matrix(y_test,y_pred))
```

# It is clear that our model is overfitted 
### Train Accuracy=98%
### Test Accuracy=77%

# Hyper-Parameter Optimization with RandomizedSearchcv

```python
from sklearn.model_selection import GridSearchCV
```

```python
#parameters
params={
"n_estimators":[x+10 for x in range(2,10)],
"min_samples_leaf":[x for x in range(2,10)],  
"min_samples_split":[x for x in range(2,10)],
"bootstrap":[True,False]
}



clf=GridSearchCV(estimator=RandomForestClassifier(),param_grid=params,cv=10,verbose=1)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.score(X_train,y_train)
```

```python
y_pred_new=clf.predict(x_test)
```

```python
print("Accuracy score(Test):",accuracy_score(y_test,y_pred_new))
```

## with a few parameters,we had reduce overfitting by 11%

### 77% accuracy is good, still you can use more parameters to gain more Accuracy,
